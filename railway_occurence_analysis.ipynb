What is the distribution of the severity levels of railway occurrences in the dataset?

```python
# Filter for occurrences with property damage or injuries
filtered_data = railDataEnglish[(railDataEnglish['AccIncTypeID_DisplayEng'] == 'Property damage') | (railDataEnglish['InjuriesIND'] == True)]

# Select relevant columns for analysis
relevant_cols = ['OccYear', 'OccurrenceTypeID_DisplayEng', 'ActivityTypeID_DisplayEng', 'TotalRSInvolved', 'TotalDGCarsInvolved', 'TotalFatalInjuries', 'TotalSeriousInjuries', 'TotalMinorInjuries', 'TotalEmployeeInjuries', 'TotalPassengerInjuries', 'TotalOtherInjuries', 'TotalMotorVehicleInjuries', 'TotalPedestrianInjuries', 'TotalTrespasserInjuries', 'NumTracks', 'VehicleDirectionID_DisplayEng', 'RoadConditionID_DisplayEng', 'RailInstalledID_DisplayEng', 'WeatherSourceID_DisplayEng', 'Visibility_Feet']

# Create a new dataframe with relevant columns only
relevant_data = filtered_data[relevant_cols]

# Plot the occurrences by year and occurrence type
plt.figure(figsize=(10, 5))
sns.countplot(x='OccYear', hue='OccurrenceTypeID_DisplayEng', data=relevant_data)
plt.title('Occurrences by Year and Occurrence Type')
plt.xlabel('Year')
plt.xticks(rotation=90)
plt.ylabel('Count')
plt.show()

# Plot the occurrences by activity type and number of tracks involved
plt.figure(figsize=(10, 5))
sns.boxplot(x='ActivityTypeID_DisplayEng', y='NumTracks', data=relevant_data)
plt.title('Occurrences by Activity Type and Number of Tracks Involved')
plt.xlabel('Activity Type')
plt.xticks(rotation=90)

plt.ylabel('Number of Tracks')
plt.show()

# Plot the occurrences by rail installation type and weather source
plt.figure(figsize=(10, 5))
sns.countplot(x='RailInstalledID_DisplayEng', hue='WeatherSourceID_DisplayEng', data=relevant_data)
plt.title('Occurrences by Rail Installation Type and Weather Source')
plt.legend(loc='upper right')
plt.xlabel('Rail Installation Type')
plt.xticks(rotation=90)

plt.ylabel('Count')
plt.show()

# Plot the occurrences by vehicle direction and road condition
plt.figure(figsize=(10, 5))
sns.countplot(x='VehicleDirectionID_DisplayEng', hue='RoadConditionID_DisplayEng', data=relevant_data)
plt.title('Occurrences by Vehicle Direction and Road Condition')
plt.xlabel('Vehicle Direction')
plt.xticks(rotation=90)

plt.ylabel('Count')
plt.show()

# Plot the distribution of injuries by type
plt.figure(figsize=(10, 5))
injury_cols = ['TotalFatalInjuries', 'TotalSeriousInjuries', 'TotalMinorInjuries', 'TotalEmployeeInjuries', 'TotalPassengerInjuries', 'TotalOtherInjuries', 'TotalMotorVehicleInjuries', 'TotalPedestrianInjuries', 'TotalTrespasserInjuries']
injury_data = relevant_data[injury_cols].sum().reset_index(name='count')
sns.barplot(x='index', y='count', data=injury_data)
plt.title('Distribution of Injuries by Type')
plt.xlabel('Injury Type')
plt.xticks(rotation=90)

plt.ylabel('Count')
plt.show()
```

```python
railDataEnglish['OccTime'] = railDataEnglish['OccTime'].fillna(0)
railDataEnglish['OccTime'] = railDataEnglish['OccTime'].astype(int)
```

```python
def time_of_day_category(time):
    if 0 <= time < 600:
        return 'Night'
    elif 600 <= time < 1200:
        return 'Morning'
    elif 1200 <= time < 1800:
        return 'Afternoon'
    else:
        return 'Evening'

railDataEnglish['OccTimeCategory'] = railDataEnglish['OccTime'].apply(time_of_day_category)
```

What is the average number of trains and train cars involved in railway occurrences? How does this vary by type of occurrence?

```python
# Calculate the average number of trains and train cars involved by occurrence type
occurrence_types = railDataEnglish["OccurrenceTypeID_DisplayEng"].unique()
results = []
for occurrence_type in occurrence_types:
    subset = railDataEnglish[railDataEnglish["OccurrenceTypeID_DisplayEng"] == occurrence_type]
    avg_trains = subset["NumberTrainsInvolved"].mean()
    avg_rs_cars = subset["TotalRSInvolved"].mean()
    avg_dg_cars = subset["TotalDGCarsInvolved"].mean()
    results.append((occurrence_type, avg_trains, avg_rs_cars, avg_dg_cars))

# Convert the results to a dataframe and display it
results_df = pd.DataFrame(results, columns=["Occurrence Type", "Avg Trains Involved", "Avg RS Cars Involved", "Avg DG Cars Involved"])
print(results_df)
```

What is the percentage of railway occurrences that involve hazardous materials or dangerous goods? How does this vary by region or type of occurrence?



```python
dg_released = railDataEnglish[railDataEnglish['DGReleasedCarsIND'] == 1]

# calculate the percentage of rows with dangerous goods released
dg_percent = len(dg_released) / len(railDataEnglish) * 100

# group the data by region and calculate the percentage for each region
dg_by_region = dg_released.groupby('RegionOfOccurrence')['DGReleasedCarsIND'].count() / railDataEnglish.groupby('RegionOfOccurrence')['OccNo'].count() * 100

# group the data by type of occurrence and calculate the percentage for each type
dg_by_occurrence = dg_released.groupby('OccurrenceTypeID_DisplayEng')['DGReleasedCarsIND'].count() / railDataEnglish.groupby('OccurrenceTypeID_DisplayEng')['OccNo'].count() * 100

# print the results
print(f"Percentage of railway occurrences involving dangerous goods: {dg_percent:.2f}%")
print("\nPercentage by region:")
print(dg_by_region)
print("\nPercentage by type of occurrence:")
print(dg_by_occurrence)
```

What is the proportion of railway occurrences that involve equipment failure or track defects? Does this vary by occurrence type or location?


```python

# create a subset of the data that includes only relevant columns
subset = railDataEnglish[['OccurrenceTypeID', 'TrackInspectedIND', 'TrackInspectSeq', 'DerailedRSIND', 'UncontrolledMovementRSIND', 'InternalRailDefectsIND']]

# calculate the proportion of occurrences that involve equipment failure or track defects by occurrence type
subset['EquipmentOrTrackFailure'] = subset[['TrackInspectedIND', 'DerailedRSIND', 'UncontrolledMovementRSIND', 'InternalRailDefectsIND']].sum(axis=1) / subset.shape[0]
proportions = subset.groupby('OccurrenceTypeID')['EquipmentOrTrackFailure'].mean().sort_values()

# plot a bar chart of the proportions
plt.barh(proportions.index, proportions)
plt.xlabel('Proportion of occurrences involving equipment or track failure')
plt.ylabel('Occurrence type')
plt.title('Proportion of railway occurrences involving equipment or track failure by occurrence type')
plt.show()
```
