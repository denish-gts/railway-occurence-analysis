```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import re


# Define the text cleaning function
def clean_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove punctuation and special characters
    text = re.sub(r'[^\w\s]', '', text)
    # Tokenize into individual words
    words = nltk.word_tokenize(text)
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    # Stem the words
    stemmer = PorterStemmer()
    words = [stemmer.stem(word) for word in words]
    # Join the words back into a string
    text = ' '.join(words)
    return text

# Clean the summary text
railDataEnglish['Summary_cleaned'] = railDataEnglish['Summary'].apply(clean_text)

# Create a bag-of-words representation of the summary text
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(railDataEnglish['Summary_cleaned'])

# Train a Naive Bayes classifier
y = railDataEnglish['CauseID']
clf = MultinomialNB()
clf.fit(X, y)

# Predict the most common cause of railway occurrences
summary = 'Train derailment due to track failure'
summary_cleaned = clean_text(summary)
X_test = vectorizer.transform([summary_cleaned])
y_pred = clf.predict(X_test)
print('Most common cause of railway occurrences: ', y_pred[0])
```

What is the average number of trains and train cars involved in railway occurrences? How does this vary by type of occurrence?

```python
# Calculate the average number of trains and train cars involved by occurrence type
occurrence_types = railDataEnglish["OccurrenceTypeID_DisplayEng"].unique()
results = []
for occurrence_type in occurrence_types:
    subset = railDataEnglish[railDataEnglish["OccurrenceTypeID_DisplayEng"] == occurrence_type]
    avg_trains = subset["NumberTrainsInvolved"].mean()
    avg_rs_cars = subset["TotalRSInvolved"].mean()
    avg_dg_cars = subset["TotalDGCarsInvolved"].mean()
    results.append((occurrence_type, avg_trains, avg_rs_cars, avg_dg_cars))

# Convert the results to a dataframe and display it
results_df = pd.DataFrame(results, columns=["Occurrence Type", "Avg Trains Involved", "Avg RS Cars Involved", "Avg DG Cars Involved"])
print(results_df)
```

```python
# subset relevant columns
subset = railDataEnglish[['OccurrenceTypeID', 'EquipmentCauseID', 'TrackCauseID', 'ProvinceID', 'RegionOfOccurrence']]

# replace NaN values with 'No Defects'
subset['EquipmentCauseID'].fillna('No Defects', inplace=True)
subset['TrackCauseID'].fillna('No Defects', inplace=True)

# create new column indicating whether an occurrence has equipment or track defects
subset['HasDefects'] = subset['EquipmentCauseID'].apply(lambda x: 'Yes' if x != 'No Defects' else 'No')
subset['HasDefects'] = subset['HasDefects'] + ' ' + subset['TrackCauseID'].apply(lambda x: 'Yes' if x != 'No Defects' else 'No')

# groupby and count occurrences by type or location, with and without defects
occtype_with_defects = subset.groupby(['OccurrenceTypeID', 'HasDefects'])['OccurrenceTypeID'].count()
province_with_defects = subset.groupby(['ProvinceID', 'HasDefects'])['ProvinceID'].count()
region_with_defects = subset.groupby(['RegionOfOccurrence', 'HasDefects'])['RegionOfOccurrence'].count()

# calculate proportions of occurrences with and without defects
occtype_proportions = occtype_with_defects.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))
province_proportions = province_with_defects.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))
region_proportions = region_with_defects.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))

# plot stacked bar chart of proportions by occurrence type
ax = occtype_proportions.unstack().plot(kind='bar', stacked=True)
ax.set_xlabel('Occurrence Type')
ax.set_ylabel('% of Occurrences')
ax.set_title('Proportion of Railway Occurrences with Equipment or Track Defects by Type')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

# plot stacked bar chart of proportions by province
ax = province_proportions.unstack().plot(kind='bar', stacked=True)
ax.set_xlabel('Province')
ax.set_ylabel('% of Occurrences')
ax.set_title('Proportion of Railway Occurrences with Equipment or Track Defects by Province')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()

# plot stacked bar chart of proportions by region
ax = region_proportions.unstack().plot(kind='bar', stacked=True)
ax.set_xlabel('Region')
```
